#Data preprocessing

```python
#Remove the useles column
data.drop('EEID',axis=1,inplace=True)
```

#Dealing with the missing value

```python
from sklearn.impute import SimpleImputer

data.isnull().sum()
#To Calculate the sum of the missing values
```

##For null categorical data

```python
from sklearn.impute import SimpleImputer

# Create an imputer for categorical data ('ex: Job Title ')
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

# Select only categorical columns using their column names
categorical_data = data[['Job Title', 'Department', 'Gender', 'Ethnicity']] # Changed this line

# Fit and transform categorical columns
data[categorical_data.columns] = imputer.fit_transform(categorical_data)
```

##For null numrical data

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
# Select only numeric columns for imputation
numeric_data = data.select_dtypes(include=[np.number])
imputer.fit(numeric_data)
data[numeric_data.columns] = imputer.transform(numeric_data)
```

##For uniqe categorical data

```python
data['Full Name'].fillna('Unknown', inplace=True)
data['Country'].fillna('Unknown', inplace=True)
data['City'].fillna('Unknown', inplace=True)
```

##For Missing Value in Hire Date

```python
# Fill missing 'Hire Date' based on the most frequent value (mode) within each 'Job Title' group
data["Hire Date"] = data.groupby("Job Title")["Hire Date"].transform(lambda x: x.fillna(x.mode()[0]))

# Now let's check for any remaining missing values in 'Hire Date'
missing_values = data["Hire Date"].isnull().sum()

# Print the remaining missing values count for 'Hire Date'
print(f"Remaining missing 'Hire Date' values: {missing_values}")
```

##For Missing Value in Exit Date

```python
import datetime
from datetime import datetime
data["Exit Date"].fillna(datetime.today().date(), inplace=True)
data.isnull().sum()
```

#Change the first 5 *rows*

```python
mode_values = data.mode().iloc[0]  # Get the first row (mode)
data.loc[:2] = mode_values.values  # Replace rows 0, 1, and 2
mode_values = data.mode().iloc[0]
data.loc[:2] = mode_values.values
data.head()
```

#Largest Salary

```python
# Calculate the Total Salary (Base Salary + Bonus)
#We add the Bounce To see the total income in the Year
data["Total Salary"] = data["Annual Salary"] + (data["Annual Salary"] * data["Bonus %"] / 100)



#sort the data by 'Total Salary' in descending order to get the highest salaries
#From large to small
top_10_data = data.sort_values(by="Total Salary", ascending=False).head(10)

# Display the top 10 rows
print(top_10_data)

```

#Group data by Department

```python
# Group data by 'Department' and calculate the average age and salary
department_stats = data.groupby('Department').agg({'Age': 'mean', 'Annual Salary': 'mean'})

# Rename the columns for better readability
department_stats = department_stats.rename(columns={'Age': 'Average Age', 'Annual Salary': 'Average Salary'})

# Display the results
print(department_stats)
```

#Group data by 'Department' and 'Ethnicity'

```python
# Group data by 'Department' and 'Ethnicity' and calculate the statistics
department_ethnicity_stats = data.groupby(['Department', 'Ethnicity']).agg(
    {'Age': ['max', 'min'], 'Annual Salary': 'median'}
)

# Rename the columns for better readability
department_ethnicity_stats.columns = ['Max Age', 'Min Age', 'Median Salary']

# Reset the index to make 'Department' and 'Ethnicity' columns
department_ethnicity_stats = department_ethnicity_stats.reset_index()

# Display the results
print(department_ethnicity_stats)
```
